```python
#●	Objective: To visualize the distribution and relationship of numerical and categorical variables in the dataset.
import pandas as pd
df=pd.read_csv(r"C:\Users\polpi\Desktop\data science\assignments\Basic stats - 1\sales_data_with_discounts.csv")
df
```

```python
numerical_df = df.select_dtypes(include=['number'])
numerical_df
```

```python
numerical_df.describe()
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import style
import scipy.stats as st
from scipy.stats import skew,norm
from scipy.stats import kurtosis
import seaborn as sns
for column in numerical_df.columns:
    plt.figure(figsize=(5,5))
    plt.hist(numerical_df[column],bins=10, edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    style.use('ggplot')
    plt.show()
```

```python
for column in numerical_df.columns:
    plt.figure(figsize=(5,5))
    sns.distplot(numerical_df[column],fit=norm)
```

```python
for column in numerical_df.columns:
    x = numerical_df[column]
    print(f'Skewness of {column} is {x.skew():.6f}')
    print(f'Kurtosis of {column} is {x.kurtosis():.6f}')
```

```python
for column in numerical_df.columns:
    data = numerical_df[column].dropna()
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    print(f'Column: {column}')
    print(f'  IQR: {IQR}')
    print(f'  Outliers:')
    if not outliers.empty:
        print(outliers.to_string(index=False))
    else:
        print('  No outliers found')
    
    plt.figure(figsize=(5,5))
    plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor='lightblue', color='black'))
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.show()
```

```python
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
categorical_columns
```

```python
category = categorical_columns.drop(['Date'])
category
```

```python
for column in category:
    plt.figure(figsize=(5, 5))  # Adjust the figure size if needed
    df[column].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f'Frequency of Categories in {column}')
    plt.xlabel(f'Category {column}')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)  # Rotate the x labels for better readability
    plt.grid(axis='y', linestyle='--')  # Optional: add grid lines for the y-axis
    plt.tight_layout()  # Adjust layout to prevent overlap
    plt.show()
```

```python
# ●	Objective II: To scale numerical variables for uniformity, improving the dataset’s suitability for analytical models.
standardized_df = numerical_df.copy()
for column in numerical_df.columns:
    mu = numerical_df[column].mean()
    sigma = numerical_df[column].std()
    standardized_df[column] = (numerical_df[column] - mu) / sigma

    print(f"Column: {column}")
    print(f"  Original Mean: {mu:.6f}")
    print(f"  Original Standard Deviation: {sigma:.6f}")
    print()
```

```python
for column in numerical_df.columns:
    plt.figure(figsize=(5, 5))
    sns.distplot(numerical_df[column], color='blue', label='Original')
    plt.title(f'Original Distribution: {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.legend()
    plt.show()
    
    plt.figure(figsize=(5, 5))
    sns.distplot(standardized_df[column], color='green', label='Standardized')
    plt.title(f'Standardized Distribution: {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.legend()
    plt.show()
```
